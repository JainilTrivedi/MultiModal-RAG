```python
def get_pdfs(links=links):
  for link in links:
    id = link.split('/')[-1]
    # print(id)
    try:
      paper = next(arxiv.Client().results(arxiv.Search(id_list=[id])))
      paper.download_pdf(dirpath="/content/drive/MyDrive/Colab Notebooks/DPS")
    except:
        print(f"multiple papers with id {id} available. Skipping this id")
    # break

```

```python
aws_region = 'us-west-2'

s3 = boto3.client('s3',region_name='us-east-2', aws_access_key_id=AWS_ACCESS_KEY, aws_secret_access_key=AWS_SECRET_KEY )
dynamodb = boto3.resource('dynamodb',region_name='us-east-2', aws_access_key_id=AWS_ACCESS_KEY, aws_secret_access_key=AWS_SECRET_KEY)
table = dynamodb.Table('paper_store')
```

## Creating Key value Pairs in AWS S3.

```python
"""
  only need to run this code when we want to save new pdf into the S3 and dynamodb.
"""
def save_pdfs(pdf_dir):

  bucket_name = 'arxiv-paper-store'

  for file_name in os.listdir(pdf_dir):
    file_path = os.path.join(pdf_dir, file_name)
    if(file_name.lower().endswith(".pdf")):
       print(f"Processing file: {file_name}")
       pdf_key = f"{file_name}"
       s3.upload_file(file_path, bucket_name, pdf_key)
       s3_url = f'https://{bucket_name}.s3.amazonaws.com/{pdf_key}'
       response = table.put_item(
            Item={
                'DocumentID': os.path.splitext(file_name)[0],  #file name
                'S3URL': s3_url,
                'Metadata': {
                    'FileName': file_name,
                    'UploadedBy': 'Jainil_1_Dec'
                }
            }
        )

       print(f"Uploaded and metadata stored for {file_name}")
    else:
       print(f"Skipped non-PDF file: {file_name}")

```

## Storing the Data in Dynamo DB.

```python
def create_dynamodb_table(table_name):
    try:
        existing_table = dynamodb.Table(table_name)
        existing_table.load()
        print(f"Table '{table_name}' already exists.")
    except ClientError as e:
        if e.response['Error']['Code'] == 'ResourceNotFoundException':
            print(f"Table '{table_name}' does not exist. Creating it...")

            table = dynamodb.create_table(
                TableName=table_name,
                KeySchema=[
                    {
                        'AttributeName': 'DocumentID',
                        'KeyType': 'HASH'
                    }
                ],
                AttributeDefinitions=[
                    {
                        'AttributeName': 'DocumentID',
                        'AttributeType': 'S'
                    }
                ],
                ProvisionedThroughput={
                    'ReadCapacityUnits': 5,
                    'WriteCapacityUnits': 5
                }
            )
            # Wait until the table is active
            table.wait_until_exists()
            print(f"Table '{table_name}' has been created.")
        else:
            raise

create_dynamodb_table("paper_store")
```
